## Ejercicio 1: Tri√°ngulo de Sierpinsky

```julia
using Plots 
# 1. Considera tres v√©rtices de un tri√°ngulo

vertices = [(0.0, 0.0), (0.5, 1.0), (1.0, 0.0)] # Arreglo de los v√©rtices, las coordenadas son tuplas.

# 2. Elige un punto arbitrario dentro de la superficie de un tri√°ngulo que es la posici√≥n actual, se escogi√≥ de preferencia un punto cerca del v√©rtice.

punto = [(rand(),rand())] # Arreglo del punto como coordenada.

# 3. Elegir un v√©rtice de manera aleatoria

v = rand(vertices)
scatter(vertices) # Ploteamos los v√©rtices

# Punto medio entre "v" (uno de los v√©rtices aleatorios) y la variable "punto"

p_m= [((punto[1][1] + v[1])/2,(punto[1][2] + v[2])/2)]

# C√≥digo base 

v = rand(vertices)
p_m= [((punto[1][1] + v[1])/2,(punto[1][2] + v[2])/2)]
punto = p_m   # 5. El punto medio es ahora la posici√≥n actual
scatter!(punto)

# Se escogi√≥ de preferencia un punto cerca del v√©rtice
```

```julia
# Se repite este c√≥digo base muchas veces y editamos el gr√°fico para quitar la legend

# Se repite el c√≥digo desde 3
v = rand(vertices)
p_m= [((punto[1][1] + v[1])/2,(punto[1][2] + v[2])/2)]
scatter!(p_m, legend = false)
punto = p_m
scatter!(p_m,legend = false)
v = rand(vertices)
p_m= [((punto[1][1] + v[1])/2,(punto[1][2] + v[2])/2)]
scatter!(p_m,legend = false, title = "Triangulo de Sierpinsky")

# Si se ejecuta muchas veces se comienza a ver la Figura del fractal
```
